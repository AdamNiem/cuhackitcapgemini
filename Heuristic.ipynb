```python
import pandas as pd
from datetime import datetime, timedelta
import os

# Define file path
file_path = 'Test/Belt 1 Test Data.xlsx'

# Get sheet names
excel_file = pd.ExcelFile(file_path)
sheet_names = excel_file.sheet_names
print(f"Excel file contains {len(sheet_names)} sheets: {sheet_names}")

# Column names to use for each sheet
actual_columns = [
    "Belt Name", 
    "Timestamp", 
    "Status",
    "Unknown",
    "Vibration Frequency",
    "Vibration Amplitude",
    "Bearing Temperature",
    "Motor Temperature",
    "Belt Load",
    "Torque",
    "Noise Levels",
    "Current and Voltage",
    "Hydraulic Pressure",
    "Belt Thickness",
    "Roller Condition"
]

# Define the expected ranges for each variable
expected_ranges_data = {
    'Vibration Frequency': {'Lower Bound': 1490, 'Upper Bound': 1510},
    'Vibration Amplitude': {'Lower Bound': 0.04, 'Upper Bound': 0.06},
    'Bearing Temperature': {'Lower Bound': 60, 'Upper Bound': 80},
    'Motor Temperature': {'Lower Bound': 80, 'Upper Bound': 100},
    'Belt Load': {'Lower Bound': 1.0, 'Upper Bound': 1.4},
    'Torque': {'Lower Bound': 280, 'Upper Bound': 320},
    'Noise Levels': {'Lower Bound': 55, 'Upper Bound': 65},
    'Current and Voltage': {'Lower Bound': 14, 'Upper Bound': 16},
    'Hydraulic Pressure': {'Lower Bound': 375, 'Upper Bound': 385},
    'Belt Thickness': {'Lower Bound': 1.5, 'Upper Bound': 1.7},
    'Roller Condition': {'Lower Bound': 65, 'Upper Bound': 100}
}

# Define the failure prediction logic
failure_prediction_logic = {
    ('Vibration Frequency', 'Current and Voltage'): 5,
    ('Bearing Temperature', 'Hydraulic Pressure'): 12,
    ('Vibration Amplitude', 'Noise Levels'): 9,
    ('Motor Temperature', 'Torque'): 7,
    ('Vibration Frequency', 'Motor Temperature'): 14,
    ('Bearing Temperature', 'Noise Levels'): 10,
    ('Vibration Amplitude', 'Belt Load'): 6,
    ('Vibration Frequency', 'Bearing Temperature'): 15,
    ('Vibration Frequency', 'Belt Load'): 8,
    ('Bearing Temperature', 'Current and Voltage'): 16
}

# Function to check if a value is out of bounds
def is_out_of_bounds(value, variable_name):
    if pd.isna(value):
        return False
    
    lower_bound = expected_ranges_data[variable_name]['Lower Bound']
    upper_bound = expected_ranges_data[variable_name]['Upper Bound']
    return value < lower_bound or value > upper_bound

# Function to find matching pair and get days to failure
def get_failure_prediction(out_of_bounds_vars):
    # Check direct matches in the failure prediction logic
    for var_pair, days in failure_prediction_logic.items():
        if var_pair[0] in out_of_bounds_vars and var_pair[1] in out_of_bounds_vars:
            return var_pair, days
    
    # If no direct match, check all possible combinations
    for i in range(len(out_of_bounds_vars)):
        for j in range(i+1, len(out_of_bounds_vars)):
            pair1 = (out_of_bounds_vars[i], out_of_bounds_vars[j])
            pair2 = (out_of_bounds_vars[j], out_of_bounds_vars[i])
            
            if pair1 in failure_prediction_logic:
                return pair1, failure_prediction_logic[pair1]
            elif pair2 in failure_prediction_logic:
                return pair2, failure_prediction_logic[pair2]
    
    # If still no match and we have at least 2 variables, use the first two with default days
    if len(out_of_bounds_vars) >= 2:
        return (out_of_bounds_vars[0], out_of_bounds_vars[1]), 10
    
    return None, None

# Function to analyze a single sheet and return results
def analyze_sheet(sheet_name):
    print(f"\n{'='*80}")
    print(f"Analyzing sheet: {sheet_name}")
    print(f"{'='*80}")
    
    # Load the data from this sheet
    df_test = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
    
    print(f"Sheet shape: {df_test.shape}")
    
    # Make sure we have at least the expected number of columns
    if df_test.shape[1] < len(actual_columns):
        print(f"Warning: Sheet {sheet_name} has fewer columns ({df_test.shape[1]}) than expected ({len(actual_columns)})")
        # Create a modified column list with only the available columns
        sheet_columns = actual_columns[:df_test.shape[1]]
        # Add generic names for any missing expected columns
        while len(sheet_columns) < len(actual_columns):
            sheet_columns.append(f"Unknown_{len(sheet_columns)}")
    else:
        sheet_columns = actual_columns.copy()
        # Add generic names for any extra columns
        while len(sheet_columns) < df_test.shape[1]:
            sheet_columns.append(f"Extra_{len(sheet_columns)}")
    
    # Apply the column names
    df_test.columns = sheet_columns[:df_test.shape[1]]
    
    # Create a list to store anomaly results for this sheet
    anomaly_results = []
    
    # Iterate through each row to find anomalies
    print(f"Analyzing rows for out-of-bounds variables...")
    for index, row in df_test.iterrows():
        # List to store out-of-bounds variables
        out_of_bounds = []
        
        # Check each variable against its expected range
        for variable in expected_ranges_data.keys():
            # Skip if variable is not in this sheet
            if variable not in df_test.columns:
                continue
                
            try:
                if is_out_of_bounds(row[variable], variable):
                    out_of_bounds.append(variable)
            except Exception as e:
                print(f"Error checking {variable} in row {index}: {e}")
        
        # If two or more variables are out of bounds
        if len(out_of_bounds) >= 2:
            # Find the matching pair and days to failure
            matching_pair, days_to_failure = get_failure_prediction(out_of_bounds)
            
            # Calculate the estimated failure date
            timestamp = row.get("Timestamp", None)
            failure_date = None
            if isinstance(timestamp, (pd.Timestamp, datetime)):
                failure_date = timestamp + timedelta(days=days_to_failure)
            
            # Store the result
            result = {
                "Sheet": sheet_name,
                "Row": index,
                "Belt Name": row.get("Belt Name", "Unknown"),
                "Timestamp": timestamp,
                "Status": row.get("Status", "Unknown"),
                "Out-of-bounds Variables": out_of_bounds,
                "Matching Pair": matching_pair,
                "Days to Failure": days_to_failure,
                "Estimated Failure Date": failure_date
            }
            
            anomaly_results.append(result)
    
    return anomaly_results

# Lists to store all results and sheet-specific results
all_anomaly_results = []
sheet_results = {}

# Process each sheet
for sheet_name in sheet_names:
    # Analyze the sheet
    results = analyze_sheet(sheet_name)
    
    # Store the results
    sheet_results[sheet_name] = results
    all_anomaly_results.extend(results)
    
    # Print a summary for this sheet
    print(f"Found {len(results)} anomalies in sheet '{sheet_name}'")

# Create directory for results if it doesn't exist
results_dir = "Anomaly_Results"
os.makedirs(results_dir, exist_ok=True)

# Export combined results to Excel
combined_df = pd.DataFrame(all_anomaly_results)
combined_file_path = os.path.join(results_dir, "All_Belt_Anomaly_Results.xlsx")
combined_df.to_excel(combined_file_path, index=False)
print(f"\nCombined results saved to {combined_file_path}")

# Export individual sheet results to separate Excel files
for sheet_name, results in sheet_results.items():
    if results:  # Only export if there are results
        sheet_df = pd.DataFrame(results)
        sheet_file_path = os.path.join(results_dir, f"{sheet_name}_Anomaly_Results.xlsx")
        sheet_df.to_excel(sheet_file_path, index=False)
        print(f"Results for sheet '{sheet_name}' saved to {sheet_file_path}")

# Print detailed report for each sheet
print("\n\nDETAILED ANOMALY REPORT BY SHEET")
print("=" * 80)

for sheet_name, results in sheet_results.items():
    print(f"\nSHEET: {sheet_name}")
    print("-" * 80)
    
    if not results:
        print("No anomalies found in this sheet.")
        continue
    
    for i, result in enumerate(results, 1):
        print(f"Anomaly #{i} (Row {result['Row']}):")
        print(f"Belt: {result['Belt Name']}")
        print(f"Timestamp: {result['Timestamp']}")
        print(f"Status: {result['Status']}")
        print(f"Out-of-bounds variables ({len(result['Out-of-bounds Variables'])}): {result['Out-of-bounds Variables']}")
        print(f"Matching pair used for prediction: {result['Matching Pair']}")
        print(f"Days to failure: {result['Days to Failure']}")
        print(f"Estimated failure date: {result['Estimated Failure Date']}")
        print("-" * 80)

print("\nAnalysis complete.")
```
